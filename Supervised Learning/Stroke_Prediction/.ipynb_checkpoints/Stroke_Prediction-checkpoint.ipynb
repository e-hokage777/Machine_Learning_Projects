# Preparing data for resampling and training

```python
## importing necessary librariess
from sklearn.model_selection import train_test_split
```

```python
## creating necessary variables
rs = 9563
```

```python
X = pd.get_dummies(df.drop("stroke", axis=1), drop_first=True)
y = df["stroke"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = rs, shuffle=True, stratify=y)
```

# Resampling data

***Remember that data has some one-hot-encoded-features***

```python
## importing necessary libraries
from imblearn.over_sampling import SMOTE, RandomOverSampler, ADASYN
from imblearn.under_sampling import RandomUnderSampler
```

```python
## creating different resamples
ros = RandomOverSampler(random_state=rs)
X_ro, y_ro = ros.fit_resample(X_train, y_train)

smote = SMOTE(random_state=rs)
X_smo, y_smo = smote.fit_resample(X_train, y_train)

adasyn = ADASYN(random_state=rs)
X_ada, y_ada = adasyn.fit_resample(X_train, y_train)

## undersampling
rus = RandomUnderSampler(random_state=rs)
X_rus, y_rus = rus.fit_resample(X_train, y_train)
```

# Training different models

```python
## importing necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost

## metrics
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay, roc_auc_score

## hyperparameter tuning classes
from sklearn.model_selection import GridSearchCV, KFold
```

```python
## creating some necessary variables
kf = KFold(n_splits = 5, shuffle=True, random_state=rs)
```

```python
## creating necessary functions
def eval(y_true, X_test, estimator):
    y_pred = estimator.predict(X_test)
    accuracy = accuracy_score(y_true, y_pred)
    precision, recall, fscore, _ = precision_recall_fscore_support(y_true, y_pred)
    
    ## printing metrics
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("fscore:", fscore)
    print("AUC:", roc_auc_score(y_true,y_pred))
    
    
    ## plotting the confusion matrix
    ConfusionMatrixDisplay.from_estimator(estimator, X_test, y_true)
    plt.show()
    
    
def eval_nn(y_true, X_test, estimator):
    y_pred = estimator.predict(X_test).round()
    accuracy = accuracy_score(y_true, y_pred)
    precision, recall, fscore, _ = precision_recall_fscore_support(y_true, y_pred)
    
    ## printing metrics
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("fscore:", fscore)
    print("AUC:", roc_auc_score(y_true,y_pred))
    
    
    ## plotting the confusion matrix
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred)
    plt.show()
```
