Dataset can be found [here](https://www.kaggle.com/datasets/trolukovich/food5k-image-dataset)

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader

from torchvision import models, transforms, datasets

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from torchvision.datasets import ImageFolder
```

```python
## creating the datasets
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = datasets.ImageFolder("data/training", transform=transform)
test_dataset = datasets.ImageFolder("data/validation", transform=transform)
```

```python
# creating train and test loader
batch_size = 64

train_loader = DataLoader(train_dataset,
                          batch_size=batch_size,
                          shuffle=True
                         )

test_loader = DataLoader(test_dataset,
                          batch_size=batch_size,
                         )
```

```python
## creating model
class FeatureModel(nn.Module):
    def __init__(self):
        super(FeatureModel, self).__init__()
        self.vgg16 = models.vgg16(weights="DEFAULT")
        self.feature_gen = self.vgg16.features
        self.pool = self.vgg16.avgpool

    def forward(self, X):
        return self.pool(self.feature_gen(X)).view(X.size(0),-1).cpu().detach().numpy()
```

```python
model = FeatureModel()
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
device
```

```python
model.to(device)
```

```python
## getting the features
train_size = len(train_dataset)
test_size = len(test_dataset)
D = 25088
```

```python
X_train = np.zeros((train_size,D))
X_test = np.zeros((test_size,D))
y_train = np.zeros((train_size,1))
y_test = np.zeros((test_size,1))
```

```python
start_index = 0
for inputs, targets in train_loader:
    batch_size = targets.size(0)
    outputs = model(inputs.to(device))
    end_index = start_index+batch_size
    X_train[start_index:end_index] = outputs
    y_train[start_index:end_index] = targets.numpy().reshape(-1,1)
    start_index += batch_size
```

```python
start_index = 0
for inputs, targets in test_loader:
    batch_size = targets.size(0)
    outputs = model(inputs.to(device))
    end_index = start_index+batch_size
    X_test[start_index:end_index] = outputs
    y_test[start_index:end_index] = targets.numpy().reshape(-1,1)
    start_index += batch_size
```

```python
## standardizing
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
y_train = y_train.flatten()
```

```python
X_test = scaler.transform(X_test)
y_test = y_test.flatten()
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
train_preds = lr.predict(X_train)
```

```python
test_preds = lr.predict(X_test)
```

```python
train_acc = np.mean(train_preds == y_train)
test_acc = np.mean(test_preds == y_test)
print(f"Train Accuracy: {train_acc}, Test Accuracy: {test_acc}")
```

### Visualizing Predictions
